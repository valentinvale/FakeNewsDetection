```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the models
models = ["BERT", "RoBERTa", "DistilBERT", "XLNet"]
accuracy = [98.0, 99.5, 97.67, 98.84]
precision = [98.0, 99.5, 97.5, 98.5]
recall = [98.0, 99.5, 96.0, 93.0]
f1_score = [97.8, 99.3, 96.6, 95.6]
training_time = [12.5, 31.5, 7, 8]  # In minutes

# Calculate efficiency for training time vs performance
# Efficiency metric = (Average of Accuracy, Precision, Recall, F1-Score) / Training Time
efficiency = [
    (np.mean([accuracy[i], precision[i], recall[i], f1_score[i]]) / training_time[i]) 
    for i in range(len(models))
]

# Normalize efficiency to a percentage scale for visualization
efficiency = [e / max(efficiency) * 100 for e in efficiency]

# Define colors and bar width
bar_width = 0.18
colors = ['#3498db', '#e67e22', '#2ecc71', '#9b59b6', '#34495e']  # Blue, Orange, Green, Purple, Dark Blue

# Set positions for bars
x = np.arange(len(models))

# Create the figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Plot bars for metrics
bars1 = ax.bar(x - bar_width*1.5, accuracy, bar_width, label='Accuracy', color=colors[0], edgecolor='black', alpha=0.9)
bars2 = ax.bar(x - bar_width/2, precision, bar_width, label='Precision', color=colors[1], edgecolor='black', alpha=0.9)
bars3 = ax.bar(x + bar_width/2, recall, bar_width, label='Recall', color=colors[2], edgecolor='black', alpha=0.9)
bars4 = ax.bar(x + bar_width*1.5, f1_score, bar_width, label='F1-Score', color=colors[3], edgecolor='black', alpha=0.9)
bars5 = ax.bar(x + bar_width*2.5, efficiency, bar_width, label='Efficiency (%)', color=colors[4], edgecolor='black', alpha=0.9)

# Add titles and labels
ax.set_title("Performance and Efficiency Comparison of Transformer Models", fontsize=18, fontweight='bold', pad=20)
ax.set_xlabel("Models", fontsize=14, labelpad=10)
ax.set_ylabel("Metrics (%) and Efficiency", fontsize=14, labelpad=10)
ax.set_xticks(x)
ax.set_xticklabels(models, fontsize=12)
ax.legend(title="Metrics", fontsize=12, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)
ax.grid(axis='y', linestyle='--', linewidth=0.6, alpha=0.7)

# Annotate each bar with values
for bars in [bars1, bars2, bars3, bars4, bars5]:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f'{yval:.1f}', ha='center', va='bottom', fontsize=10, color='black')

# Modern aesthetic improvements
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()

# Show the plot
plt.show()
```
